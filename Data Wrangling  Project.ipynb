```python

#Treating for outliers

def outlier_lims(col):
 q3,q1 = np.percentile(col, [75,25])
 iqr = q3-q1
 upper_lim = q3 + 1.5*iqr
 lower_lim = q1 - 1.5*iqr
 return upper_lim, lower_lim
```

```python
for col in continous_vars:
 print("--------------------------------------------------")
 print("Column:", col)
 
 UL,LL = outlier_lims(df_New[col])
 print("Upper Limit =", UL)
 print("Lower Limit =", LL)
 
 total_outliers = len(df_New.loc[df_New[col]<LL,col]) + len(df_New.loc[df_New[col]>UL,col])
 percent = (total_outliers / len(df_New.index) )*100
 
 print("Percentage of Outliers=", percent)
 print("-------------------------------------------------- \n")
```
